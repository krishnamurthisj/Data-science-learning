```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('sl1.csv')
```

```python
df
```

```python
df.head()
```

```python
target=df[['Price']]
features=df.drop(columns=['Price'])
```

```python
x_train,x_test,y_train,y_test=train_test_split(features,target,train_size=0.7,random_state=10)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
####model tarining
lin_model=LinearRegression()
lin_model.fit(x_train,y_train)
```

```python
b0=lin_model.intercept_
b0
```

```python
b1=lin_model.coef_
b1
```

```python
b0+b1*3600
```

```python
####training validation
y_pred=lin_model.predict(x_train)
r2_score(y_train,y_pred)
```

```python
y_pred1=lin_model.predict(x_test)

r2_score(y_test,y_pred1)
```

```python
sns.regplot(x=y_test,y=y_pred1,color='red',line_kws={'color':'blue'})
plt.show()
```
